```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from bokeh.plotting import output_file
from sklearn.preprocessing import LabelEncoder
from matplotlib import patches
```

```python
data=pd.read_csv('starbucks.csv')
data
```

```python
#Vissualizacion 1
print('Vissualizacion 1')
fig, axs = plt.subplots(1, figsize=(5, 5))
sns.histplot(data=data['Calories'], ax=axs, color='blue', alpha=0.5, label='Calories')
sns.histplot(data=data[' Sodium (mg)'], ax=axs, color='red', alpha=0.5, label='Sodium (mg)')
plt.legend()
```

```python
#Vissualizacion 2
print('Vissualizacion 2')
data["Beverage_prep"].value_counts().plot(kind='pie')
```

```python
#Vissualizacion 3
print('Vissualizacion 3')
plt.figure(figsize=(20,6))
sns.boxplot(data=data, x='Beverage_category', y=' Dietary Fibre (g)')
plt.title('Caregoria de Bebida vs Fibra Dietetica')
plt.show()
```

```python
#Vissualizacion 4
print('Vissualizacion 4')
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Eliminamos los valores nulos
dataNotNull = data.dropna(subset=['Saturated Fat (g)', 'Trans Fat (g) ', ' Sugars (g)'])

le = LabelEncoder()
data['Beverage_category'] = le.fit_transform(data['Beverage_category'])

# Creamos el gráfico de dispersión
p = ax.scatter(data["Saturated Fat (g)"], data["Trans Fat (g) "] , data[" Sugars (g)"], c=data["Beverage_category"])
ax.set_xlabel('Grasa Saturada')
ax.set_ylabel('Grasas Trans')
ax.set_zlabel('Azucar')
plt.title('Grasas y Azucar en Bebidas de Starbucks')
fig.colorbar(p)
plt.show()
```

```python
#Vissualizacion 5
print('Vissualizacion 5')
fig, axs = plt.subplots(1, figsize=(5, 5))
sns.scatterplot(data=data, x='Calories', y=' Total Fat (g)', ax=axs)
```

```python
#Vissualizacion 6
print('Vissualizacion 6')
plt.figure(figsize=(15,6))
sns.boxplot(data=data, x='Beverage_category', y='Caffeine (mg)')
plt.title('Caregoria de Bebida vs Fibra Dietetica')
plt.show()
```

```python
#Vissualizacion 7
print('Vissualizacion 7')
vitamin_a_counts = data["Vitamin A (% DV) "].value_counts()
vitamin_c_counts = data["Vitamin C (% DV)"].value_counts()
plt.figure()
plt.plot(vitamin_a_counts.index, vitamin_a_counts.values, label='Vitamin A')
plt.plot(vitamin_c_counts.index, vitamin_c_counts.values, label='Vitamin C')
plt.legend()
plt.show()
```

```python
#Vissualizacion 8
print('Vissualizacion 8')
fig, axs = plt.subplots(1, figsize=(5, 5))
sns.boxplot(data=data, y=' Total Carbohydrates (g) ', ax=axs)
```

```python
#Vissualizacion 9
print('Vissualizacion 9')
data1 = data[[" Protein (g) "]] 
plt.figure(figsize=(10,5))
sns.heatmap(data1)
plt.title('Gráfico de calor')
plt.show()
#
```
